# FonksiyonlarÄ±n Ã‡Ä±ktÄ±sÄ±: Return

YukarÄ±da Ã¶ÄŸrencilerin not ortalamasÄ±nÄ± bulduk, ekrana yazdÄ±rabildik fakat <strong>kullanamadÄ±k</strong>. Ã–ÄŸrenciye sadece not ortalamasÄ±nÄ± yazdÄ±rmak yerine hangi harf notunu aldÄ±ÄŸÄ±nÄ± gÃ¶steren bir program yazmak isteseydik ne yapacaktÄ±k ğŸ¤”

TÄ±pkÄ± daha Ã¶nce kullandÄ±ÄŸÄ±mÄ±z <strong>len()</strong> fonksiyonu gibi bizim fonksiyonumuz da istediÄŸimiz bilgiyi ekrana yazmak yerine kodumuzun iÃ§erisinde kullanmamÄ±z iÃ§in bize verse?

Ä°ÅŸte bu sebeple fonksiyonun yaptÄ±ÄŸÄ± iÅŸlemin sonucunu kodumuzun devamÄ±nda kullanacaksak **return** kullanmalÄ±yÄ±z. **return** kodun Ã§Ä±ktÄ±sÄ±nÄ± bir <strong><font color="orange">deÄŸiÅŸkende tutabilmemizi</font></strong> saÄŸlar.


````python
def ortalama_hesapla(vize, final, vize_yÃ¼zde=40, final_yÃ¼zde=60):
  ortalama = (vize * vize_yÃ¼zde / 100) + (final * final_yÃ¼zde / 100)
  return ortalama
````

````python
ortalama_bilgisi = ortalama_hesapla(70, 80)
print(f"OrtalamanÄ±z: {ortalama_bilgisi}")

# bu ortalama_bilgisi deÄŸiÅŸkenini dilediÄŸimiz ÅŸekilde kullanabiliriz:
if ortalama_bilgisi >= 90:
  print("Harf notunuz A, harikasÄ±nÄ±z!")
elif ortalama_bilgisi >= 75:
  print("Harf notunuz B, tebrikler!")
elif ortalama_bilgisi >= 65:
  print("Harf notunuz C, bir dahakine biraz daha gayret :)")
elif ortalama_bilgisi >= 50:
  print("Harf notunuz D, ucuz atlattÄ±n! :)")
else:
  print("ÃœzgÃ¼nÃ¼m, dersten kaldÄ±nÄ±z :/")
````

Åimdi tÃ¼m Ã¶ÄŸrencilere harf notu bilgisini verelim:

````python
notlar = [['Ali', 20, 41], ['Serap', 89, 48], ['YÄ±ldÄ±z', 75, 85], ['Fatih', 83, 66], ['Lale', 90, 95]]
for puanlar in notlar:
  ortalama = ortalama_hesapla(puanlar[1], puanlar[2])
  print(f"Sevgili {puanlar[0]}, ortalamanÄ±z {ortalama:.2f}")
  if ortalama >= 90:
    print("Harf notunuz A, harikasÄ±nÄ±z!")
  elif ortalama >= 80:
    print("Harf notunuz B, tebrikler!")
  elif ortalama >= 70:
    print("Harf notunuz C, bir dahakine biraz daha gayret :)")
  elif ortalama >= 60:
    print("Harf notunuz D, ucuz atlattÄ±n! :)")
  else:
    print("ÃœzgÃ¼nÃ¼m, dersten kaldÄ±nÄ±z :/")
  
  print('-'*20)
````

# Ã‡eÅŸitli Return KullanÄ±mlarÄ±

YukarÄ±daki harf notu hesaplayan kodumuz epey kullanÄ±ÅŸlÄ±, onu da bir fonksiyona dÃ¶nÃ¼ÅŸtÃ¼rsek?

````python
def harf_notu_hesapla(ortalama):
  if ortalama >= 90:
    return 'A'
  elif ortalama >= 75:
    return 'B'
  elif ortalama >= 65:
    return 'C'
  elif ortalama >= 50:
    return 'D'
  else:
    return 'F'
````

````python
# ortalama_hesaplama kodumuzu hatÄ±rlayalÄ±m
def ortalama_hesapla(vize, final, vize_yÃ¼zde=40, final_yÃ¼zde=60):
  ortalama = (vize * vize_yÃ¼zde / 100) + (final * final_yÃ¼zde / 100)
  return ortalama
````

BÃ¶ylelikle iki bilgiyi de dilediÄŸimiz gibi kullanabilir, hatta <strong> ortalama_hesapla()</strong> fonksiyonundan elde ettiÄŸimiz sayÄ±yÄ± <strong>harf_notu_hesapla()</strong> fonksiyonuna verebiliriz:

````python
ortalama_puan = ortalama_hesapla(90,70)
harf_notu = harf_notu_hesapla(ortalama_puan)
print(f"OrtalamanÄ±z: {ortalama_puan}, harf notunuz: {harf_notu}")
````

````python
notlar = [['Ali', 20, 41], ['Serap', 89, 48], ['YÄ±ldÄ±z', 75, 85], ['Fatih', 83, 66], ['Lale', 90, 95]]
for puanlar in notlar:
  ortalama_puan = ortalama_hesapla(puanlar[1], puanlar[2])
  harf_notu = harf_notu_hesapla(ortalama_puan)
  print(f"Sevgili {puanlar[0]}, ortalaman: {ortalama_puan:.2f}, harf notun {harf_notu}")
````

## Skimage

Biraz da gÃ¶rsellerle uÄŸraÅŸalÄ±m ğŸŒ 

skimage kÃ¼tÃ¼phanesinden fotoÄŸraflara eriÅŸelim:

````python
import skimage
kedi = skimage.data.cat()
print(kedi)
````

Ah, bu pek beklediÄŸimiz bir ÅŸey deÄŸildi ğŸ™„ GÃ¶rÃ¼ntÃ¼leyebilmek iÃ§in bir baÅŸka kÃ¼tÃ¼phaneye ihtiyacÄ±mÄ±z var: matplotlib

````python
import matplotlib.pyplot as plt
plt.imshow(kedi, cmap='gray')        
plt.axis('off')    
plt.show()
````

Bundan sonra kolaylÄ±k olmasÄ± aÃ§Ä±sÄ±ndan gÃ¶rÃ¼ntÃ¼leme fonksiyonu oluÅŸturalÄ±m:

````python
import matplotlib.pyplot as plt
def gÃ¶rÃ¼ntÃ¼le(resim):
  plt.imshow(resim, cmap='gray')        
  plt.axis('off')    
  plt.show()
````

````python
gÃ¶rÃ¼ntÃ¼le(kedi)
````

````python
import numpy
kedi_dÃ¶ndÃ¼r = numpy.fliplr(kedi)
gÃ¶rÃ¼ntÃ¼le(kedi_dÃ¶ndÃ¼r)
````

Åimdi biraz da filtreler ğŸ“·

````python
from skimage.color import rgb2gray

# Kedi resmini siyah beyaz yapalÄ±m
gri_kedi = rgb2gray(kedi)

# Orijinal resim
gÃ¶rÃ¼ntÃ¼le(kedi)

# Siyah beyaz resim
gÃ¶rÃ¼ntÃ¼le(gri_kedi)
````

https://scikit-image.org/docs/stable/api/skimage.data.html

````python
from skimage.filters import gaussian
gaussian_kedi = gaussian(kedi, multichannel=True)
gÃ¶rÃ¼ntÃ¼le(gaussian_kedi)
````

````python
gaussian_kedi = gaussian(kedi, sigma=5, multichannel=True)
gÃ¶rÃ¼ntÃ¼le(gaussian_kedi)
````

````python
from skimage.filters import sobel

# Apply edge detection filter
edge_sobel = sobel(gri_kedi)

# Show original and resulting image to compare
gÃ¶rÃ¼ntÃ¼le(kedi)
gÃ¶rÃ¼ntÃ¼le(edge_sobel)
````

## NLTK

NLTK temel dil iÅŸleme kÃ¼tÃ¼phanelerinden biri.

````python
import nltk # natural language toolkit
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
````

````python
sentence = 'The Autobiography of Malcolm X was published in 1965, the result of a collaboration between human rights activist Malcolm X and journalist Alex Haley.'
print(nltk.word_tokenize(sentence))
````

NLTK ve Wikipedia kÃ¼tÃ¼phanelerini beraber kullanabiliriz. Ã–ncelikle Ä°stanbul sayfasÄ±nda geÃ§en kelimeleri bir listeye dÃ¶nÃ¼ÅŸtÃ¼relim

````python
istanbul = wikipedia.page("Ä°stanbul").content
kelimeler = nltk.word_tokenize(istanbul)
print(kelimeler)
````

Ä°stanbul kelimesi kaÃ§ kere geÃ§miÅŸ bulalÄ±m:

````python
kelimeler.count('Istanbul')
````

KaÃ§ farklÄ± kelime olduÄŸunu bulalÄ±m, fakat Ã¶nce listedeki parantez, virgÃ¼l gibi kelime olmayan karakterleri temizleyelim.

````python
sÃ¶zlÃ¼k = []
for kelime in kelimeler:
  if kelime in "#$%&\'()*+,-./:;<==>?@[\\]^_`{|}~":
    continue

  if kelime not in sÃ¶zlÃ¼k:
    sÃ¶zlÃ¼k.append(kelime)

print(sÃ¶zlÃ¼k)
````
