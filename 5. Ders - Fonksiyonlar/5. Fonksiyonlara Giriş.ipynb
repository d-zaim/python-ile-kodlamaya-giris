Hmm, demek ki bir parametre tanımlamışsak daima onun için bir değer (argument) vermeliyiz. Peki oldu ki kullanıcı ismini öğrenemedik ama yine de bu fonksiyonu çalıştırmamız gerekiyor. O halde **default argument** kullanacağız:

````python
def karşılama(isim='kıymetli misafirimiz'):
  print(f"Hoş geldiniz {isim},")
  print('Harika bir gösteri sizi bekliyor :)')
````

````python
karşılama()
````

````python
karşılama('Ali')
````

Problem çözüldü, şahane! 

İlla tek bir parametre mi yazmak zorundayız? Elbette HAYIR 

````python
def karşılama(isim, soyad):
  print(f"Hoş geldiniz sayın {isim} {soyad},")
  print('Harika bir gösteri sizi bekliyor :)')
````

````python
x = 'ali'
y = 'yıldırım'
karşılama(y, x)
````

Bu tanımladığımız fonksiyonu bir arkadaşımıza gönderdik, arkadaşımız ilk başta isim yerine soyad, ikinci olarak da isim girmiş ve garip bir şeyle karşılaşmış:

````python
karşılama('Yıldırım','Ali')
````

Demek ki parametreler için verdiğimiz değerlerin (argument) __konum__ları önemliymiş! Buna **positional argument** deniyor. E ama ben her yazılan fonksiyonda hangi parametre kaçıncı nereden bileceğim? İşte bunu bilmenin imkansızlığından dolayı şu şekilde de fonksiyonları kullanabiliyoruz.

````python
karşılama(soyad='Yıldırım', isim='Ali')
````

Bu şekilde kullanıma **keyword argument** deniyor. 

Burada dikkat etmemiz gereken şey, bir kere keyword argument kullandıysak, sonrasında positional argument **kullanamayız.** Yoksa aşağıdaki hatayı alırız:

````python
karşılama(soyad='Yıldırım','Ali')
````

Neticede 3 farklı şekilde parametre kullanabiliriz:
1. Positional Argument
2. Default Argument
3. Keyword Argument

## Hesap Makinesi

````python
def toplama(x,y):
  print(x+y)

def çıkarma(x,y):
  print(x-y)

def çarpma(x,y):
  print(x*y)

def bölme(x,y):
  print(x/y)
````

````python
while(True):
  işlem = input('Yapmak istediğiniz işlemi seçiniz: (+) (-) (*) (/)')
  sayı_1 = int(input('İlk sayıyı giriniz: '))
  sayı_2 = int(input('İkinci sayıyı giriniz: '))

  if işlem == '+':
    toplama(sayı_1, sayı_2)

  elif işlem == '-':
    çıkarma(sayı_1, sayı_2)

  elif işlem == '*':
    çarpma(sayı_1, sayı_2)

  elif işlem == '/':
    bölme(sayı_1, sayı_2)

  else:
    print("Lütfen belirtilen 4 işlemden birini seçiniz.")
    continue
    
  çıkış = input('Devam etmek için evet yazınız, istemiyorsanız boş bırakabilirsiniz: ')
  if çıkış != 'evet':
    print('Güle güle :)')
    break
````

# Fonksiyonların Çıktısı: Return

Şu ana kadar yaptığımız işlemlerin çıktısı için hep print fonksiyonunu kullandık. Fakat print kullandığımızda fonksiyonun çıktısını sadece **görebiliyoruz, kullanamıyoruz.** Fonksiyonun yaptığı işlemin sonucunu kodumuzun devamında kullanacaksak **return** kullanmalıyız. **return** kodun çıktısını bir değişkende tutabilmemizi sağlar.


````python
def ortalama_hesapla(vize, final, vize_yüzde=40, final_yüzde=60):
  ortalama = (vize * vize_yüzde / 100) + (final * final_yüzde / 100)
  return ortalama
````

````python
notlar = [[20, 41], [59, 48], [70, 35], [83, 37], [53, 95]]
for puanlar in notlar:
  ortalama_hesapla(puanlar[0], puanlar[1])
````

Hiçbir şey göremedik? Çünkü return kod yazan bize bu bilgiyi kodun devamında kullanmamız için verir, ekrana yazdırmaz. Tıpkı **input** fonksiyonu, **len** fonksiyonu gibi.

````python
isimler = ['Ali','Esra','Ayşe','Zübeyir']
for isim in isimler:
  len(isim)
````

````python
isimler = ['Ali','Esra','Ayşe','Zübeyir']
for isim in isimler:
  uzunluk = len(isim)
  print(f"{isim}, isminiz {uzunluk} harflidir.")
````

````python
notlar = [['Ali', 20, 41], ['Serap', 59, 48], ['Yıldız', 70, 35], ['Fatih', 83, 37], ['Lale', 53, 95]]
for puanlar in notlar:
  x = ortalama_hesapla(puanlar[1], puanlar[2])
  print(f"Sevgili {puanlar[0]}, ortalaman {x}")
````

# Çeşitli Return Kullanımları

````python
def ortalama_hesapla(vize, final, vize_yüzde=40, final_yüzde=60):
  ortalama = (vize * vize_yüzde / 100) + (final * final_yüzde / 100)
  
  if ortalama > 90:
    return 'A'

  elif ortalama > 80:
    return 'B'

  elif ortalama > 70:
    return 'C'
    
  elif ortalama > 60:
    return 'D'

  else:
    return 'F'
````

````python
harf_notu = ortalama_hesapla(20,40)
print(harf_notu)
````

````python
notlar = [['Ali', 20, 41], ['Serap', 59, 48], ['Yıldız', 70, 35], ['Fatih', 83, 37], ['Lale', 53, 95]]
for puanlar in notlar:
  harf = ortalama_hesapla(puanlar[1], puanlar[2])
  print(f"Sevgili {puanlar[0]}, harf notun {harf}")
````

## Skimage

````python
from skimage import data
kedi = data.cat()
kedi
````

````python
import matplotlib.pyplot as plt
def görüntüle(resim):
  plt.imshow(resim, cmap='gray')        
  plt.axis('off')    
  plt.show()
````

````python
kedi = data.cat()
görüntüle(kedi)
````

````python
kedi_döndür = np.fliplr(data.cat())
görüntüle(kedi_döndür)
````

````python
from skimage.color import rgb2gray

# Kedi resmini siyah beyaz yapalım
gri_kedi = rgb2gray(kedi)

# Orijinal resim
görüntüle(kedi)

# Siyah beyaz resim
görüntüle(gri_kedi)
````

https://scikit-image.org/docs/stable/api/skimage.data.html

````python
from skimage.filters import gaussian
gaussian_kedi = gaussian(kedi, multichannel=True)
görüntüle(gaussian_kedi)
````

````python
gaussian_kedi = gaussian(kedi, sigma=5, multichannel=True)
görüntüle(gaussian_kedi)
````

````python
from skimage.filters import sobel

# Apply edge detection filter
edge_sobel = sobel(gri_kedi)

# Show original and resulting image to compare
görüntüle(kedi)
görüntüle(edge_sobel)
````
