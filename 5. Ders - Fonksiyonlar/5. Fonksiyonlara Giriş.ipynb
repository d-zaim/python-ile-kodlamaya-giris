# Fonksiyonların Çıktısı: Return

Yukarıda öğrencilerin not ortalamasını bulduk, ekrana yazdırabildik fakat <strong>kullanamadık</strong>. Öğrenciye sadece not ortalamasını yazdırmak yerine hangi harf notunu aldığını gösteren bir program yazmak isteseydik ne yapacaktık 🤔

Tıpkı daha önce kullandığımız <strong>len()</strong> fonksiyonu gibi bizim fonksiyonumuz da istediğimiz bilgiyi ekrana yazmak yerine kodumuzun içerisinde kullanmamız için bize verse?

İşte bu sebeple fonksiyonun yaptığı işlemin sonucunu kodumuzun devamında kullanacaksak **return** kullanmalıyız. **return** kodun çıktısını bir <strong><font color="orange">değişkende tutabilmemizi</font></strong> sağlar.


````python
def ortalama_hesapla(vize, final, vize_yüzde=40, final_yüzde=60):
  ortalama = (vize * vize_yüzde / 100) + (final * final_yüzde / 100)
  return ortalama
````

````python
ortalama_bilgisi = ortalama_hesapla(70, 80)
print(f"Ortalamanız: {ortalama_bilgisi}")

# bu ortalama_bilgisi değişkenini dilediğimiz şekilde kullanabiliriz:
if ortalama_bilgisi >= 90:
  print("Harf notunuz A, harikasınız!")
elif ortalama_bilgisi >= 75:
  print("Harf notunuz B, tebrikler!")
elif ortalama_bilgisi >= 65:
  print("Harf notunuz C, bir dahakine biraz daha gayret :)")
elif ortalama_bilgisi >= 50:
  print("Harf notunuz D, ucuz atlattın! :)")
else:
  print("Üzgünüm, dersten kaldınız :/")
````

Şimdi tüm öğrencilere harf notu bilgisini verelim:

````python
notlar = [['Ali', 20, 41], ['Serap', 89, 48], ['Yıldız', 75, 85], ['Fatih', 83, 66], ['Lale', 90, 95]]
for puanlar in notlar:
  ortalama = ortalama_hesapla(puanlar[1], puanlar[2])
  print(f"Sevgili {puanlar[0]}, ortalamanız {ortalama:.2f}")
  if ortalama >= 90:
    print("Harf notunuz A, harikasınız!")
  elif ortalama >= 80:
    print("Harf notunuz B, tebrikler!")
  elif ortalama >= 70:
    print("Harf notunuz C, bir dahakine biraz daha gayret :)")
  elif ortalama >= 60:
    print("Harf notunuz D, ucuz atlattın! :)")
  else:
    print("Üzgünüm, dersten kaldınız :/")
  
  print('-'*20)
````

<i>`ortalama:.2f` kullanımı ondalıklı sayılarda virgülden sonra kaç basamak göstermesini istediğimizi belirtiyor. </i>

# Çeşitli Return Kullanımları

Yukarıdaki harf notu hesaplayan kodumuz epey kullanışlı, onu da bir fonksiyona dönüştürsek?

````python
def harf_notu_hesapla(ortalama):
  if ortalama >= 90:
    return 'A'
  elif ortalama >= 75:
    return 'B'
  elif ortalama >= 65:
    return 'C'
  elif ortalama >= 50:
    return 'D'
  else:
    return 'F'
````

````python
# ortalama_hesaplama kodumuzu hatırlayalım
def ortalama_hesapla(vize, final, vize_yüzde=40, final_yüzde=60):
  ortalama = (vize * vize_yüzde / 100) + (final * final_yüzde / 100)
  return ortalama
````

Böylelikle iki bilgiyi de dilediğimiz gibi kullanabilir, hatta <strong> ortalama_hesapla()</strong> fonksiyonundan elde ettiğimiz sayıyı <strong>harf_notu_hesapla()</strong> fonksiyonuna verebiliriz:

````python
ortalama_puan = ortalama_hesapla(90,70)
harf_notu = harf_notu_hesapla(ortalama_puan)
print(f"Ortalamanız: {ortalama_puan}, harf notunuz: {harf_notu}")
````

````python
notlar = [['Ali', 20, 41], ['Serap', 89, 48], ['Yıldız', 75, 85], ['Fatih', 83, 66], ['Lale', 90, 95]]
for puanlar in notlar:
  ortalama_puan = ortalama_hesapla(puanlar[1], puanlar[2])
  harf_notu = harf_notu_hesapla(ortalama_puan)
  print(f"Sevgili {puanlar[0]}, ortalaman: {ortalama_puan:.2f}, harf notun {harf_notu}")
````

Böylelikle Ali'den beklentilerimizi ayrıca dile getirebiliriz:

````python
notlar = [['Ali', 20, 41], ['Serap', 89, 48], ['Yıldız', 75, 85], ['Fatih', 83, 66], ['Lale', 90, 95]]
for puanlar in notlar:
  ortalama_puan = ortalama_hesapla(puanlar[1], puanlar[2])
  harf_notu = harf_notu_hesapla(ortalama_puan)
  
  if puanlar[0] == 'Ali' and ortalama_puan < 70:
    print(f"Ali, not ortalaman {ortalama_puan:.2f} ve harf notun {harf_notu}")
    print("Kodlamayla uğraşmaktan derslerini ihmal ediyorsun gibi görünüyor 🤨")
  else:
    print(f"Sevgili {puanlar[0]}, ortalaman: {ortalama_puan:.2f}, harf notun {harf_notu}")
  
  print('-'*40)
````

````python
import skimage
kedi = skimage.data.cat()
print(kedi)
````

Ah, bu pek beklediğimiz bir şey değildi 🙄 Görüntüleyebilmek için bir başka kütüphaneye ihtiyacımız var: matplotlib

````python
import matplotlib.pyplot as plt
plt.imshow(kedi, cmap='gray')        
plt.axis('off')    
plt.show()
````

Bundan sonra kolaylık olması açısından görüntüleme fonksiyonu oluşturalım:

````python
import matplotlib.pyplot as plt
def görüntüle(resim):
  plt.imshow(resim, cmap='gray')        
  plt.axis('off')    
  plt.show()
````

````python
görüntüle(kedi)
````

````python
import numpy
kedi_döndür = numpy.fliplr(kedi)
görüntüle(kedi_döndür)
````

Şimdi biraz da filtreler 📷

````python
from skimage.color import rgb2gray

# Kedi resmini siyah beyaz yapalım
gri_kedi = rgb2gray(kedi)

# Orijinal resim
görüntüle(kedi)

# Siyah beyaz resim
görüntüle(gri_kedi)
````

https://scikit-image.org/docs/stable/api/skimage.data.html

````python
from skimage.filters import gaussian
gaussian_kedi = gaussian(kedi, multichannel=True)
görüntüle(gaussian_kedi)
````

````python
gaussian_kedi = gaussian(kedi, sigma=5, multichannel=True)
görüntüle(gaussian_kedi)
````

## NLTK

NLTK temel dil işleme kütüphanelerinden biri.

````python
import nltk # natural language toolkit
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
````

````python
sentence = 'The Autobiography of Malcolm X was published in 1965, the result of a collaboration between human rights activist Malcolm X and journalist Alex Haley.'
print(nltk.word_tokenize(sentence))
````

NLTK ve Wikipedia kütüphanelerini beraber kullanabiliriz. Öncelikle İstanbul sayfasında geçen kelimeleri bir listeye dönüştürelim

````python
istanbul = wikipedia.page("İstanbul").content
kelimeler = nltk.word_tokenize(istanbul)
print(kelimeler)
````

İstanbul kelimesi kaç kere geçmiş bulalım:

````python
kelimeler.count('Istanbul')
````

Kaç farklı kelime olduğunu bulalım, fakat önce listedeki parantez, virgül gibi kelime olmayan karakterleri temizleyelim.

````python
sözlük = []
for kelime in kelimeler:
  if kelime in "#$%&\'()*+,-./:;<==>?@[\\]^_`{|}~":
    continue

  if kelime not in sözlük:
    sözlük.append(kelime)

print(sözlük)
````

Epey temizlendi, peki son olarak hangi kelimenin kaç kere geçtiğini bulabilir misiniz?
