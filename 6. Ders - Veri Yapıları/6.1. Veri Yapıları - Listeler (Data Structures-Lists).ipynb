#### 10.1. .append() Metodu

```python
# sublist -> alt liste / liste içinde liste / listenin alt listesi
liste = [1,2,3,4,5,6]
liste.append(["a","b","c"])
liste
```

```python
liste1 = [1,2,3,4,5,6]
liste3= ["a","b","c"]
liste2 = []
for eleman in liste1:    # liste1'deki her bir elemanı al:
    print("hangi elemandayız:" , eleman)
    print("liste2'nin bir önceki döngüdeki hali:" , liste2)
    liste2.append(eleman**2)
    print("liste2'nin yeni hali:", liste2)
    print("------")
```

### 12.Listelerde Döngüler

#### index ve elemana aynı anda erişme: enumerate()

```python
liste = ['elma','armut','portakal','mandalina','erik']
for eleman in liste:
  print(eleman)
```

```python
for index in range(len(liste)):
  print(index+1,liste[index])
```

```python
liste = ['elma','armut','portakal','mandalina','erik']
for index, eleman in enumerate(liste):
  print(f"{index}: {eleman}")
```

```python
liste = ['elma','armut','portakal','mandalina','erik']
for index, eleman in enumerate(liste, 1):
  print(f"{index}: {eleman}")
```

#### Birden fazla listede döngü yazma: zip()

```python
# Ülkeler ve başkentleri eşleştirelim:
ülkeler = ['Türkiye','Almanya','İngiltere']
başkentler = ['Ankara','Berlin','Londra']

zip(ülkeler,başkentler)
```

```python
list(zip(ülkeler,başkentler))
```

```python
filmler = ['The Shawshank Redemption', 'The Godfather', 'The Godfather: Part II', 'The Dark Knight', '12 Angry Men',
           "Schindler's List", "The Lord of the Rings: The Return of the King", "Pulp Fiction",
           "The Good, the Bad and the Ugly","The Lord of the Rings: The Fellowship of the Ring"]
tarihler = [1994, 1972, 1974, 2008, 1957, 1993, 2003, 1994, 1966, 2001]
for index in range(len(filmler)):
  print(filmler[index],tarihler[index])
```

```python
filmler = ['The Shawshank Redemption', 'The Godfather', 'The Godfather: Part II', 'The Dark Knight', '12 Angry Men',
           "Schindler's List", "The Lord of the Rings: The Return of the King", "Pulp Fiction",
           "The Good, the Bad and the Ugly","The Lord of the Rings: The Fellowship of the Ring"]
tarihler = [1994, 1972, 1974, 2008, 1957, 1993, 2003, 1994, 1966, 2001]

for film, tarih in zip(filmler, tarihler):
  print(f"{film} - {tarih}")
```

```python
filmler = ['The Shawshank Redemption', 'The Godfather', 'The Godfather: Part II', 'The Dark Knight', '12 Angry Men',
           "Schindler's List", "The Lord of the Rings: The Return of the King", "Pulp Fiction",
           "The Good, the Bad and the Ugly","The Lord of the Rings: The Fellowship of the Ring"]
tarihler = [1994, 1972, 1974, 2008, 1957, 1993, 2003, 1994, 1966, 2001]

for index, (film, tarih) in enumerate(zip(filmler, tarihler), 1):
  print(f"{index}. {film} - {tarih}")
```
