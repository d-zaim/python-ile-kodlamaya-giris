## 4.2. Kümelerde Operatörler

### 1) Birleşim (Union)

```python
küme1 = set(['çilek','kiraz'])
küme2 = set(['çilek','şeftali'])
küme1.union(küme2) 
```

```python
# .union()'a alternatif:
küme1 | küme2 
```

```python
# birden fazla küme birleştirme:
küme3 = set(['erik','üzüm'])
küme1.union(küme2,küme3)
```

### 2) Kesişim (Intersection)

```python
küme1 = set(['çilek','kiraz'])
küme2 = set(['çilek','şeftali'])
küme1.intersection(küme2)
```

```python
orijinal_gruplar = ['A','AB','B','0']
```

```python
hasta_kan_grupları = ['ABB', 'ABB', 'Z', 'B', 'B', 'ABB', 'AB', '0', 'ABB', 'Z', 'B', 'ABB', 'ABB', '0', 'B', 'A', 'Z', 'AB', '0', 'ABB']
set(hasta_kan_grupları)
```

```python
set(hasta_kan_grupları).intersection(set(orijinal_gruplar))
```

```python
# .intersection()'a alternatif:
küme1&küme2
```

### 3) Fark (Difference)

```python
# Sadece küme1'de olanlar
küme1.difference(küme2)
```

```python
küme1 - küme2
```

```python
# sadece hasta kan grupları listesinde olan elemanlar
set(hasta_kan_grupları).difference(set(orijinal_gruplar))
```

```python
# sadece orijinal kan grupları listesinde olan elemanlar
set(orijinal_gruplar).difference(set(hasta_kan_grupları))
```

### 4) Disjoint (Ayrık kümeler mi?)

```python
küme1.isdisjoint(küme3-{'b'})
```

```python
küme1.intersection(küme3)
```

```python
# ayrık kümelerse intersectionları boş küme olur:
küme1.intersection(küme3-{'b'})
```

### 5) Subset (Alt kümesi mi?)

```python
küme1 = {1,2,3,4,5}
küme2 = {1,2,3}
küme2.issubset(küme1)
```

```python
küme2 <= küme1
```

### 6) Superset (Üst kümesi mi?)

```python
küme1 >= küme2
```

```python
küme1.issuperset(küme2)
```

## 4.3. Kümelerde Metodlar

### 1) add()

```python
küme1 = {1,2,3,4,5}
küme1.add(6)
küme1
```

### 2) remove()

```python
küme1.remove(6)
küme1
```

```python
# Eğer olmayan eleman için remove fonksiyonunu kullanırsak hata verir:
küme1.remove(10)
```

### 3) discard()

```python
# Olmayan elemanı yok etmek istediğimizde discard hata vermez:
küme1.discard(10)
```

### 4) pop()

```python
# Elemanı kümeden atar:
küme1 = {1,2,3,4,5}
küme1.pop()
küme1
```
