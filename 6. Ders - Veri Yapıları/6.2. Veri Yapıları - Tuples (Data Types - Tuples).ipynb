### 2.2. Packing-Unpacking (Paketleme-Paketi Açma)

```python
eşyalar = ('kapı','dolap','sandalye','masa','lamba')
eşya_1 = eşyalar[0]
eşya_2 = eşyalar[1]
```

```python
eşya_1, eşya_2, eşya_3, eşya_4, eşya_5 = 'kapı','dolap','sandalye','masa','lamba'
print(eşya_1)
print(eşya_2)
print(eşya_3)
print(eşya_4)
print(eşya_5)
```

```python
a,b = 1,2
print(a)
print(b)
```

Peki ya eksik yazarsak?

```python
eşyalar = ('kapı','dolap','sandalye','masa','lamba')
eşya_1, eşya_2 = eşyalar
```

```python
eşyalar = ('kapı','dolap','sandalye','masa','lamba')
eşya_1, *eşya_2 = eşyalar
print(eşya_1)
print(eşya_2)
```

```python
eşya_1, *eşya_2, eşya_3 = eşyalar
print(eşya_1)
print(eşya_2)
print(eşya_3)
```

Tuple oluştururken () şart değildir. Virgül konulması yeterli:

```python
ali, ayse, zehra = 19, 22, 35
ali
```

Ali ile Ayşe'nin yaşlarını değiştirmek isteseydik ne yapmamız gerekiyordu? (swapping variable values)

```python
ali, ayse = 19, 22
print('Ali', ali)
print('Ayşe', ayse)
geçici_değişken = ali
ali = ayse
ayse = geçici_değişken 
print('Ali', ali)
print('Ayşe', ayse)
```

Tuple'ların unpacking özelliği sayesinde bunu tek satırda yapabiliriz:

```python
ali, ayse = 19, 22
ali, ayse = ayse, ali
ali
```

Unpacking özelliğini for döngüleri içerisinde kullanabiliriz.

```python
yazarlar = [('Sait Faik', 'Abasıyanık', 1954), ('Bilge', 'Karasu', 1995), ('Vüsat','Bener',2005)]

for yazar in yazarlar:
    print(f"{yazar} yılında vefat etmiştir.")
```

```python
yazarlar = [('Sait Faik', 'Abasıyanık', 1954), ('Bilge', 'Karasu', 1995), ('Vüsat','Bener',2005)]

for ad, soyad, vefat in yazarlar:
    print(f"{ad} {soyad} {vefat} yılında vefat etmiştir.")
```

```python
for ad, soyad in yazarlar:
    print(f"{ad} {soyad}")
```

Sadece ad soyad kısmını kullanmak isteseydik, vefat yerine "\_" atayabilirdik.
_alt tire mi? O da nereden çıktı?_ dediyseniz şayet, zaten bu tip değişkenlere "dummy variable" deniyor 🤭

```python
yazarlar = [('Sait Faik', 'Abasıyanık', 1954), ('Bilge', 'Karasu', 1995), ('Vüsat','Bener',2005)]

for ad, soyad, _ in yazarlar:
    print(f"{ad} {soyad} Türk edebiyatının önemli isimlerinden birisidir.")
```

Bu dummy variable'ı kullanmasaydık olabilecekleri tahmin edebiliyorsunuz:


```python
yazarlar = [('Sait Faik', 'Abasıyanık', 1954), ('Bilge', 'Karasu', 1995), ('Vüsat','Bener',2005)]

for ad, soyad in yazarlar:
    print(f"{ad} {soyad} Türk edebiyatının önemli isimlerinden birisidir.")
```

### 2.3. Packing-unpacking'den *args'lara!
Daha önce yazdığımız fonksiyonlarda argüman sayımız hep kısıtlıydı. Toplama fonksiyonu yazdık sadece 2 eleman alabildik 😞 Artık durum değişti!

Sayısını bilmediğim kadar argümanın hepsini al anlamında: *args kullanabiliriz.


```python
değişken1, *değişken2 = 1,2,3,'a','b'
değişken2
```

```python
def bomba(x, *patates): 
  print(x)
  print(patates)
```

```python
bomba(1,2,3,'a','b')
```

```python
def bomba(*patates):
  print(patates)
  for eleman in patates:
    print(eleman)
```

```python
bomba('a',1,2)
```

ve tabii illa *args demek şart değil. *args sadece genel bir kullanım.

```python
def toplama(*sayılar):
  toplam = 0
  for eleman in sayılar:
    toplam += eleman
  return toplam
```

```python
toplama(1,2,3,4,5,6,7)
```

### 2.4. Tuplelar ile İşlemler

İki tuple birbirine listelerdeki gibi eklenebilir.

```python
şehirler = ('İstanbul', 'Ankara')
şehirler_2 = ('Bursa','İzmir')

bütün_şehirler = şehirler + şehirler_2
bütün_şehirler
```

Tuple içinde hızlıca arama yapmak için **in** kullanabiliriz.

```python
şehirler = ('İstanbul', 'Ankara')
print("İstanbul" in şehirler)
```

Aradığımız değerin hangi indexte olduğunu bulabiliriz.

```python
şehirler = ('İstanbul', 'Ankara','Manisa','Erzurum','Kars')
print(şehirler.index('Erzurum'))
```

İç içe tuplelar:

```python
şehirler = ('İstanbul', 'Ankara')
şehirler_2 = ('Bursa','İzmir')

bütün_şehirler = (şehirler, şehirler_2)
bütün_şehirler
```

```python
# Ülkeler ve başkentleri eşleştirelim:
ülkeler = ('Türkiye','Almanya','İngiltere')
başkentler = ('Ankara','Berlin','Londra')

zip(ülkeler,başkentler)
```

```python
for i in range(len(ülkeler)):
  print(ülkeler[i], başkentler[i])
```

```python
for ülke, başkent in zip(ülkeler,başkentler):
  print(ülke.upper(), başkent)
```
