## 3.1. Dictionaries Giriş

Sözlükler, şu ana kadar gördüğümüz listeler ve tuplelara göre çok farklı ve güzel bir özelliğe sahip. Esasen bildiğimiz *telefon rehberi* ya da *sözlük*. Listelerin içinde bulunan verilere index üzerinden erişiyorduk. Ama index üzerinden gittiğimiz için verilerin ne *anlama geldiği* bilgisini tutamıyorduk. Dictionary'ler sayesinde bilgiye *bizim için anlamlı olan* bir keyword vereceğiz. Bu keyword'ü sözlükte yazdığımızda aradığımız bilgiye ulaşacağız. Özet olarak aşağıdaki gibi bir tablo şeklinde de düşünülebilir.

| Key (Ürün Adı)   | Value (Ücret) |
| -----------      | -----------   |
| Çanta              | 265 |
| Ayakkabı        | 320 |
| Şapka          | 85   |

## 3.4. Dictionary Metodları

### get()

```python
ürün_ücretleri = {
    'Çanta':265,
    'Ayakkabı':320,
    'Şapka':85
}
ürün_listesi = ['Çanta','Ayakkabı','Şapka']
for ürün in ürün_listesi:
  print(f"{ürün} ücreti: {ürün_ücretleri[ürün]}")
```

Peki ya listede sözlükte olmayan bir ürün yazsaydı?

```python
ürün_listesi = ['Çanta','Ayakkabı','Kazak','Şapka']
for ürün in ürün_listesi:
  print(f"{ürün} ücreti: {ürün_ücretleri[ürün]}")
```

```python
ürün_ücretleri['Kazak']
```

**KeyError** aldık, yani böyle bir key değerinin sözlüğümüzde mevcut olmadığını söylüyor. Döngünün çökmemesi için sözlükte olmayan kelimeleri aramayalım

```python
'Kazak' in ürün_ücretleri
```

```python
'Çanta' in ürün_ücretleri
```

```python
ürün_listesi = ['Çanta','Ayakkabı','Şapka','Atkı']
for ürün in ürün_listesi:
  if ürün in ürün_ücretleri:
    print(f"{ürün} ücreti: {ürün_ücretleri[ürün]}")
  else:
    print(f"{ürün} sistemimizde kayıtlı değildir.")
```

İşte sözlüklerde olmayan değeri tespit için *if-else* yerine kullanabileceğimiz bir metot var: **get()** metodu.

```python
ürün_ücretleri['Ayakkabı']
```

```python
ürün_ücretleri.get('Ayakkabı')
```

```python
ürün_ücretleri
```

```python
ürün_ücretleri.get('Atkı')
```

```python
ürün_listesi = ['Çanta','Ayakkabı','Şapka','Atkı']
for ürün in ürün_listesi:
    print(f"{ürün} ücreti: {ürün_ücretleri.get(ürün)}")
```

Olmayan bir **key** değeri girdiğimizde **KeyError** yazmak yerine en azından hiçbir şey vermedi. Fakat burada da bir bilgilendirme yok. Onun için **get** fonksiyonunun **default** parametresini kullanabiliriz.

```python
ürün_ücretleri.get('Atkı','bilinmeyen ürün')
```

Harika! Döngüyü yeniden düzenleyelim:

```python
ürün_listesi = ['Çanta','Ayakkabı','Şapka','Atkı']
for ürün in ürün_listesi:
  print(f"{ürün} ücreti: {ürün_ücretleri.get(ürün,'bilinmiyor')}")
```

### update()

Aynı listelerdeki extend gibi, dictionary'lerde de update metodu var.

```python
ürün_ücretleri = {
    'Çanta':265,
    'Ayakkabı':320,
    'Şapka':85
}
yeni_ürünler = {
    "Kazak":95,
    "Gömlek": 63,
    "Kemer":40
}
ürün_ücretleri.update(yeni_ürünler)
ürün_ücretleri
```

```python
liste1 = [1,2,3]
liste2 = [4,5,6]
liste1.append(liste2)
liste1
```

```python
liste1 = [1,2,3]
liste2 = [4,5,6]
liste1.extend(liste2)
liste1
```

## 3.5. If-else Alternatifi Olarak Dictionary Kullanmak

```python
# hesap makinesi örneği
def toplama(x,y):
  return x+y
def çıkarma(x,y):
  return x-y
def çarpma(x,y):
  return x*y
def bölme(x,y):
  return x/y
```

```python
işlem = input("yapmak istediğiniz işlemi seçiniz: (+)/(-)/(*)/(/) ")
sayı1 = int(input('ilk sayı: '))
sayı2 = int(input('ikinci sayı: '))
hesap_makinesi = {
    "+": toplama,
    "-": çıkarma,
    "*": çarpma,
    "/": bölme
}

#toplama(sayı1, sayı2)
hesap_makinesi.get(işlem)(sayı1, sayı2)
```
