## 4) Sınıf Eylemleri (Methods)

__Metodlar__, oluşturduğumuz objelerin yapabildiği eylemler. Yani? Yani oluşturduğumuz her bir obje
- Kendisi üzerinde değişiklik yapabilir (araba hızlanabilir)
- Kendi özelliklerine göre bir çıktı verebilir (araba kendi hızına göre belirli bir mesafeyi ne kadar gidebileceğini veya ne kadar yakıt harcayacağını söyleyebilir.)
- Tıpkı bir fonksiyon gibi belirli bir işi yapabilir.

Ve benzeri çeşitli şekillerde eylemlere (__methods__) sahip olabilir.

Kendisi üzerinde değişikliği daha önce epey kullandık. string metodlarının önemli bir kısmı böyleydi:

````python
metin = 'bir metin belgesi'
metin.upper()
````

Daha önce metod ve fonksiyon ayrımını işte bu yüzden yapmıştık. Metodlar __objelerin fonksiyonları__dır ve __obje üzerinden çağırılırlar__. 

metin.title()

kullanımındaki noktanın sebebi bu. metin objesinin yani esasen string objesinin title eylemini çağırıyoruz ve bunu kendini değiştirerek gerçekleştiriyor.

Kendi özelliklerine göre bir çıktı (return) vermesini de yine stringler üzerinden hatırlayabiliriz:

````python
metin.count('e')
````

````python
metin_2 = 'bu metinde kaç eeee var'
metin_2.count('e')
````

Şimdi bu metodları kendimiz yazalım.

````python
class Araba():
  # Attributes
  renk = 'mavi'
  güç = 250
  hız = 50 

  # Methods
  def hız_göster(self):
    print(self.hız)

  def hız_değiştir(self, yeni_hız):
    self.hız = yeni_hız

  def süre_hesapla(self, yol):
    süre = yol / self.hız
    print(f"Bu araba {yol} km yolu {süre} saatte gidebilir.")
````

Burada dikkat etmemiz gereken şey, metodların __her zaman en az 1 parametresi olur__ ve o da __self__'tir. __self__ objenin kendisi demektir. O halde self.hız dediğimizde oluşturulmuş _bir_ araba objesinin _kendi_ hızı anlamına geliyor. Birkaç araba oluşturup görelim.

## 5) Sınıfların Kurucu Metodları (Constructors)

Yukarıda işlemimizi uzatan bir şeyler oldu. Arabaları oluşturduk ama mecburen 2 araba da aynı özelliklere sahip doğdu ve biz buna en baştan müdahale edemedik. Daha sonra __hız_değiştir__ metoduyla değiştirdik. Bunun yerine araba objesinin özelliklerini __daha oluşturulma aşamasında__ belirleyebiliriz.

Objenin yapısına ait değişiklik yapan bu tip metodlara __constructors__ veya __magic methods__ veya __dunder methods__ deniyor (dunder: duble underscore). Bu şekilde söylenmesinin sebebi kurucu metod isimlerinin iki alt tire arasına yazılmasından dolayı.

Aşağıda en temel kurucu metodlardan __\_\_init\_\___ metodunu görelim. Bu metod __initialize__ kelimesinin kısaltılmışı. Her bir obje doğduğunda/oluşturulduğunda (initialize) bu fonksiyon çalışacak.

````python
class Araba():
  def __init__(self, hız=150, marka='Mercedes', renk='mavi'):
    self.hız = hız
    self.marka = marka
    self.renk = renk
  
  def süre_hesapla(self, yol):
    süre = yol / self.hız
    print(f"Bu araba {yol} km yolu {süre} saatte gidebilir.")
````

Burada bizim hiç yazmadığımız bir sürü çift alt tireli (dunder) metod var?? İşte bu Object oriented programming'in bir ileri aşaması. Objelerin de kendi aralarında hiyerarşik bir yapı var. Bir obje özelliklerinin ve metodlarının bir kısmını başka bir objeden _miras_ alabilir. 

Burada karşımıza çıkan ve bizim hiç yazmadığımız metodlar aslında __en üst obje__den miras geliyor. Her obje (tanımlanmasa dahi) bu metodlarla doğuyor gibi düşünebiliriz. Bunlardan birini kullanalım.

Şu ana kadar bir çok veri tipi arasında _eşitlik_ bağlantısı kurduk. İşte bu __\_\_eq\_\___ dunder metoduyla yapılıyor:

````python
ilk = [1,2,3]
ikinci = [1,2,3]
ilk == ikinci
````

````python
# Bu ikisi aynı obje midir?
ilk is ikinci
````

````python
# sıralamaya göre eşitlik değişiyordu. Bu tamamen bu şekilde tanımlandığı için.
[1,2,3] == [1,3,2]
````

````python
# Kendi Araba sınıfımızda "iki araba arasında eşit olma durumu"nu tanımlayalım
class Araba():
  def __init__(self, hız=150, marka='Mercedes', renk='mavi'):
    self.hız = hız
    self.marka = marka
    self.renk = renk
  
  def süre_hesapla(self, yol):
    süre = yol / self.hız
    print(f"Bu araba {yol} km yolu {süre} saatte gidebilir.")


  # Eğer iki arabanın hızları birbirine eşitse o arabalar birbirine eşittir.
  def __eq__(self, other):
    if (self.hız == other.hız) and (self.marka == other.marka):
      return True
    else:
      return False

  def __repr__(self):
    return f'Bu {self.hız} km/h hızında, {self.renk} renginde {self.marka} bir arabadır.'
````

````python
araba1 = Araba()
araba2 = Araba(250, 'BMW','sarı')
araba3 = Araba(250, 'Toyota', 'kırmızı')
````

````python
araba1 == araba2
````

````python
araba2 == araba3
````

````python
araba2 is araba3
````

````python
print(araba1)
print(araba2)
print(araba3)
````

## 3) Sklearn
Machine learning kütüphanesi.

Kod kaynağı:https://www.datacamp.com/community/tutorials/svm-classification-scikit-learn-python

````python
#Import scikit-learn dataset library
from sklearn import datasets

#Load dataset
cancer = datasets.load_breast_cancer()
````

````python
# print the names of the 13 features
print("Features: ", cancer.feature_names)
````

````python
# print the label type of cancer('malignant' 'benign')
print("Labels: ", cancer.target_names)
````

````python
# print data(feature)shape
cancer.data.shape
````

````python
# print the cancer data features (top 5 records)
print(cancer.data[0:5])
````

````python
# print the cancer labels (0:malignant, 1:benign)
print(cancer.target)
````

````python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    test_size=0.3,random_state=109) # 70% training and 30% test
````

````python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
````

````python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
````

````python
# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred))
````

## 5) Spacy
Sözel verilerle çalışmak (Natural Language Processing) için oluşturulmuş bir kütüphane.

Ücretsiz online kursu: https://course.spacy.io/en/

````python
import spacy

# Create a blank English nlp object
nlp = spacy.blank("en")

# Created by processing a string of text with the nlp object
doc = nlp("Hello world!")

# Iterate over tokens in a Doc
for token in doc:
    print(token.text)
````

````python
doc = nlp("It costs $5.")
print("Index:   ", [token.i for token in doc])
print("Text:    ", [token.text for token in doc])

print("is_alpha:", [token.is_alpha for token in doc])
print("is_punct:", [token.is_punct for token in doc])
print("like_num:", [token.like_num for token in doc])
````

````python
# Load the small English pipeline
nlp = spacy.load("en_core_web_sm")

# Process a text
doc = nlp("She ate the pizza")

# Iterate over the tokens
for token in doc:
    # Print the text and the predicted part-of-speech tag
    print(token.text, token.pos_)
````

````python
# Process a text
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
# Iterate over the predicted entities
for ent in doc.ents:
    # Print the entity text and its label
    print(ent.text, ent.label_)
````
