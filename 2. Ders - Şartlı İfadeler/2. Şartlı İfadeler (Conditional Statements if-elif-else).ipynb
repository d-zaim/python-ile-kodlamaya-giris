### Değişkenler
Python yazdığımız her bilgiyi (veriyi) hafızasında tutuyor. Bu hafızada tuttuğu yerin adresini biz bilmiyoruz, bilmemiz de iyi olmazdı ("Ali" bilgisi için şöyle bir adresi _139848550483440_ ezberlemek istemezdik 🙄). Bunun için biz __o adresteki bilgiye erişmek için bir etiket__ belirliyoruz ve bu etiketlemeye teknik olarak __değişken (variable)__ deniyor.

_Not: Değişkenin tuttuğu bilginin adresine id() fonksiyonu ile bakabiliriz._

````python
isim = 'Ali'
print(id(isim))
````

Değişken kullanmanın bir avantajı bu, bir diğer avantajı şu:

Bir bilgi birkaç yerde kullanılacaksa değişken olarak atamak en doğrusu. Neden? Çünkü insanoğlunun tabiatı tembel 🤫 3-5 değişiklik yapmaktansa 1 değişiklik yapmayı tercih ederiz. İşte bir örnek:

````python
hesap1 = 5000
hesap2 = 10000
hesap1_dolar = 5000 / 13.5
hesap2_dolar = 10000 / 13.5

print('Bugünkü dolar kuru: ' + str(13.5))
print("Hesap_1 özetiniz: " + str(hesap1))
print("Hesap_1 dolar: " + str(hesap1_dolar))
print("Hesap_2 özetiniz: " + str(hesap2))
print("Hesap_2 dolar: " + str(hesap2_dolar))
````

Ve evet korktuğunuz şey oldu, dolar kuru değişti: artık 13.4. Kaç tane değiştirmeniz gereken yer var? 3 😐

Dolar kurunu bir __değişken__ olarak kaydetseydik kaç yerde değişiklik yapmamız gerekirdi? 1 😏

````python
dolar_kuru = 13.5
hesap1 = 5000
hesap2 = 10000
hesap1_dolar = 5000 / dolar_kuru
hesap2_dolar = 10000 / dolar_kuru

print('Bugünkü dolar kuru: ' + str(dolar_kuru))
print("Hesap_1 özetiniz: " + str(hesap1))
print("Hesap_1 dolar: " + str(hesap1_dolar))
print("Hesap_2 özetiniz: " + str(hesap2))
print("Hesap_2 dolar: " + str(hesap2_dolar))
````

```
import random
```

 Bundan sonra random kütüphanesindeki "*rastgele bir integer üret*" anlamındaki **randint** fonksiyonunu kullanmak için şu şekilde yazacağız:
