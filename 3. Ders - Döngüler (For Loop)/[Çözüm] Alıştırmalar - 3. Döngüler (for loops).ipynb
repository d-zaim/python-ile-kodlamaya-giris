## 9\. Alıştırma

Bir tekstil fabrikasında üretilen kazaklar paketlenmek için bir bant ile taşınıyor. Bu bantın üzerine bir kamera yerleştirilmiş, her ürünün rengini tek tek bir listeye kaydediliyor. Banttaki bütün kazakların rengi aynı olmalı. Bantta farklı renk olan kazağı bulan bir program yazalım.
```
bant = ['siyah','siyah','siyah','kırmızı','siyah','siyah']
```
Örnek çıktı:
```
4. kazak hatalı, lütfen düzeltiniz.
```


````python
bant = ['siyah','siyah','siyah','kırmızı','siyah','siyah']

doğru_renk = ''
doğru_renk_sayısı = 0

for renk in bant:
  if bant.count(renk) > doğru_renk_sayısı:
    doğru_renk = renk
    doğru_renk_sayısı = bant.count(renk)


for index in range(len(bant)):
  if bant[index] != doğru_renk:
    print(f"{index+1}. kazak hatalı, lütfen düzeltiniz.")
````

__ÖNEMLİ NOT:__ Tek bir kod ile hem bant2 hem bant3'te tespit yapabilmelisiniz. Dolasıyla hangi rengin esas, hangi rengin hatalı olduğunu bulmanız gerekiyor. Renkleri doğrudan tanımlayarak yapmayacağız. 

Hatalı örnek:
```
if renk != 'siyah':

```
Bu şekilde yazarsak bant3'teki esas renk turuncu olduğu için kodunuz bant3'te doğru çalışmaz.




````python
bant2 = ['siyah','siyah','siyah','mavi','siyah','siyah','mavi','siyah','siyah','mavi','siyah']
bant3 = ['turuncu','mor','turuncu','turuncu','mor','turuncu','turuncu','mor','turuncu','turuncu','turuncu','mor']
hatalılar = []
````

````python
# 1. Çözüm
doğru_renk = ''
doğru_renk_sayısı = 0
hatalılar = []

for renk in bant2:
  if bant2.count(renk) > doğru_renk_sayısı:
    doğru_renk = renk
    doğru_renk_sayısı = bant2.count(renk)


for index in range(len(bant2)):
  if bant2[index] != doğru_renk:
    hatalılar.append(index+1)

print(f"{hatalılar} numaralı kazaklar hatalı, lütfen düzeltiniz.")
````

````python
# 2. Çözüm
bant3 = ['turuncu','mor','turuncu','turuncu','mor','turuncu','turuncu','mor','turuncu','turuncu','turuncu','mor']
renk1 = []
renk2 = []

for i in range(len(bant3)):
  if bant3[0] == bant3[i]:
    renk1.append(i+1)
  else:
    renk2.append(i+1)
    
# çok olan renk doğru renktir.
if len(renk1) > len(renk2):
  print(f"{renk2} numaralı kazaklar hatalı, lütfen düzeltiniz.")
else:
  print(f"{renk1} numaralı kazaklar hatalı, lütfen düzeltiniz.")
````

Google'ın güçlü şifre önermesi gibi biz de 8 haneli büyük harf, küçük harf ve sayı içeren rastgele bir şifre üreten bir program yazalım.

Alfabedeki tüm harfler için aşağıdaki string'i kullanabilirsiniz.


```
harfler = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
sayılar = '0123456789'
```


Bunun için birkaç aşamaya ihtiyacımız var:
1. Bu stringlerden rastgele bir eleman seçelim (random.choice() fonksiyonuna daha önce liste yazmıştık, string de yazabiliriz.)
2. Her döngüde rastgele oluşan harfi veya sayıyı yeni bir string'e sırayla ekleyelim.
3. Kullanıcının güçlü şifresini ekrana yazdıralım.

````python
import random
harfler_küçük = 'abcdefghijklmnopqrstuvwxyz'
harfler_büyük = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
sayılar = '0123456789'
hepsi = harfler_küçük + harfler_büyük + sayılar
şifre = ''

rastgeleler = [random.randint(0,7),random.randint(0,7),random.randint(0,7)]
print(rastgeleler)

for i in range(8):
  if i == rastgeleler[0]:
    şifre += random.choice(sayılar)
  elif i == rastgeleler[1]:
    şifre += random.choice(harfler_küçük)
  elif i == rastgeleler[2]:
    şifre += random.choice(harfler_büyük)
  else:
    şifre += random.choice(hepsi)

print(şifre)
````
