Bunu √∂nlemek i√ßin if-else kullanabiliriz. Eƒüer puan + 5, 100'√º ge√ßiyorsa o puanƒ± 100'e tamamlayalƒ±m, diƒüerlerine +5 ekleyelim.

````python
puan_listesi = [42,33,51,16,23,99,93,58,67,92,48,36,75,84,97]
yeni_puan_listesi = []
for puan in puan_listesi:
  yeni_puan_listesi.append(puan+5)

print(yeni_puan_listesi)
````

````python
puan_listesi = [42,33,51,16,23,99,93,58,67,92,48,36,75,84,97]
yeni_puan_listesi = []

for puan in puan_listesi:
  if (puan+5) > 100:
    puan = 100
  else:
    puan += 5
  yeni_puan_listesi.append(puan)

print(puan_listesi)
print(yeni_puan_listesi)
````

````python
# Alternatif
puan_listesi = [42,33,51,16,23,99,93,58,67,92,48,36,75,84,97]
yeni_puan_listesi = []
for puan in puan_listesi:
  puan += 5
  if puan > 100:
    puan = 100
  yeni_puan_listesi.append(puan)

print(puan_listesi)
print(yeni_puan_listesi)
````
