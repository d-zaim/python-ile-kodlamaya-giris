````python
# Döngüyle listedeki her bir puanı 5 artırıp yeni listeye ekleyelim.
# Orijinal puan_listesi hiç değişmeyecek.

puan_listesi = [42,33,51,16,23,91,93,58,67,92,48,36,75,84]
yeni_puan_listesi = []

for puan in puan_listesi:
  yeni_puan_listesi.append(puan+5)
  print(yeni_puan_listesi)

print('Döngü bitti ')
print(puan_listesi)
print(yeni_puan_listesi)
````

Peki hangisini seçelim? Listeyle ilgili yapmak istediğimiz işleme bağlı.

1. Listedeki elemanların __yeri önemli değilse__ doğrudan liste üzerinde for loop yapabiliriz: 
```
for puan in puan_listesi:
  ...
```
 Örneğin:
 - listedeki puanların toplamını bulma.
 - Listede ortalamanın üzerinde alınan puanları bulma.
 - Listedeki meyvelerden birinin sayısını bulma. ("elma kaç kere geçiyor?" gibi.)
 
2. Listedeki elemanların __yeri önemli ise__ _range_ kullanabiliriz.
```
for index in range(5):
  ...
```
Örneğin:
  - Listedeki elemanların sıralamasını göstermemiz gerekiyorsa:
    1. The Shawshank Redemption
    2. The Godfather 
  - Listedeki elemanların _yerlerini_ bulmamız gerekiyorsa ("elma 1. 3. ve 7. sırada bulunuyor" gibi bir bilgiyi elde etmek için)
  - 2 listeyi beraber kullanacaksak (notumuzun en sonunda mevcut)

### 3) if-else ile Birleştirelim

İTİRAF: Biraz önceki örnekte matematik öğretmeninin listesi biraz hileliydi. Çünkü listedeki hiçbir öğrencinin notu 5 ile toplanınca 100'ü geçmiyordu 🙄 Ya geçseydi?

Bunu önlemek için if-else kullanabiliriz. Eğer puan + 5, 100'ü geçiyorsa o puanı 100'e tamamlayalım, diğerlerine +5 ekleyelim.

````python
puan_listesi = [42,33,51,16,23,99,93,58,67,92,48,36,75,84,97]
yeni_puan_listesi = []
for puan in puan_listesi:
  yeni_puan_listesi.append(puan+5)

print(yeni_puan_listesi)
````

````python
puan_listesi = [42,33,51,16,23,99,93,58,67,92,48,36,75,84,97]
yeni_puan_listesi = []

for puan in puan_listesi:
  if (puan+5) > 100:
    puan = 100
  else:
    puan += 5
  yeni_puan_listesi.append(puan)

print(puan_listesi)
print(yeni_puan_listesi)
````

````python
# Alternatif
puan_listesi = [42,33,51,16,23,99,93,58,67,92,48,36,75,84,97]
yeni_puan_listesi = []
for puan in puan_listesi:
  puan += 5
  if puan > 100:
    puan = 100
  yeni_puan_listesi.append(puan)

print(puan_listesi)
print(yeni_puan_listesi)
````

### 5) İki Farklı Listede Döngü

Uzunlukları **tamamen aynı** olan 2 listede döngü yazabiliriz.

````python
filmler = ['The Shawshank Redemption', 'The Godfather', 'The Godfather: Part II', 'The Dark Knight', '12 Angry Men',
           "Schindler's List", "The Lord of the Rings: The Return of the King", "Pulp Fiction",
           "The Good, the Bad and the Ugly","The Lord of the Rings: The Fellowship of the Ring"]
tarihler = [1994, 1972, 1974, 2008, 1957, 1993, 2003, 1994, 1966, 2001]
````

````python
len(filmler)
````

````python
len(tarihler)
````

````python
print(f"{filmler[0]} filmi {tarihler[0]} tarihinde vizyona çıkmıştır.")
print(f"{filmler[1]} filmi {tarihler[1]} tarihinde vizyona çıkmıştır.")
````

````python
liste_uzunluk = len(filmler)
for i in range(liste_uzunluk):
  print(f"{filmler[i]} filmi {tarihler[i]} tarihinde vizyona çıkmıştır.")
````

````python
# Genelde bu şekilde yazılır:

for i in range(len(filmler)):
  print(f"{filmler[i]} filmi {tarihler[i]} tarihinde vizyona çıkmıştır.")
````

IMDB ilk 10'daki 2000 sonrası filmleri bulalım.

Burada range kullanıyoruz, çünkü tarihler listesindeki 2000'den büyük her bir tarihin __index__'ini alıyoruz. Bu __index__'i kullanıp filmler listesinde aynı __index__'teki filmi buluyoruz.

Böylece tarihler listesinde belirli bir filtreleme yaptık, bu bilgiyle filmler listesinden istediğimiz filtreye göre filmleri seçmiş olduk.



````python
ikibinliler = []
for i in range(len(filmler)):
  if tarihler[i] > 2000:
    ikibinliler.append(filmler[i])

print(ikibinliler)
````
